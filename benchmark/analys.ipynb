```python
import json
```

```python
task_name = "cifar100_cnum100_dist1_skew0.8_seed0"
with open(f"fedtask/{task_name}/data.json", "r") as f:
    data = json.load(f)
list_label_train = []
list_label_valid = []
for c in data['client_names']:
    label_train = label = data[c]['dtrain']['y']
    label_valid = label = data[c]['dvalid']['y']
    list_label_train.append([label_train.count(l) for l in range(100)])
    list_label_valid.append([label_valid.count(l) for l in range(100)])
```

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array(list_label_train)
plt.imshow(arr,)
```

```python
arr = np.array(list_label_valid)
plt.imshow(arr,)
```
